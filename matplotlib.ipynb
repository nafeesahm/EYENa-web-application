# LinePlot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt              # pyplot use for extra  like colour, design or x, y

# %matplotlib inline                         # magical line , use in colab
```

```python
#Graph
```

```python
no_of_class= np.array([1,2,3,4,5,6,7])
no_of_student= np.array([4,9,6,4,10,8,6])
plt.plot(no_of_class, no_of_student)
# plt.show()                                                    
```

```python
plt.plot(no_of_class, no_of_student, c='g' , LineStyle = '-.')                #c= colour                 # shift + tab   for cheet sheet
```

```python
plt.plot(no_of_class, no_of_student , linewidth= 5)
plt.xlabel('no_of_class')
plt.ylabel('no_of_student')
plt.title('Attendence', c='g')           # c='g' means colour is green
```

# Barchart

```python
subject= ['AI', 'ML', 'DBMS', 'SQL']
m1= [5,50,12,14]
m2= [12,14,20,5]
m3= [10,18,30,20]
```

```python
plt.bar(subject,m1, width= .5 , color= 'r')
```

```python
plt.bar(subject, m2 , align= 'center')
```

```python
plt.bar(subject, m3, width= .8 , align= 'edge')
```

# ML and DBMS  both class are attend equal ( 50 ,  20+30= 50 )

# --------------------------------------------------------------------

```python
plt.barh(subject, m3 , align= 'edge')             # horizontal bar
```

# pieChart

```python
a= np.array([32,35,78,90])
myLabel=['india','America','USA','Canada']
mycolor= ['red', 'green', 'yellow', 'black']
plt.pie(a, labels = myLabel, colors= mycolor, radius= 2)              # by default radius = 1 
#plt.show()
```

# subplot

```python

s = np.array([231,672,893,560])
d = np.array([67,98,12,77])
plt.plot(s,d)
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(1,2,1)                        # subplot(1,2,1) .... ( nrows , ncol , index , **kwargs)    (x_axis ,y_axis)
plt.plot(x,y)                                                                   # index is important
```

```python

a =np.array([3,4,8,6])
b = np.array([2,9,7,5])
plt.subplot(2,1,2)
plt.plot(a,b)
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(1,2,2)                        # multiple plot same index 
plt.plot(x,y)

a =np.array([3,4,8,6])
b = np.array([2,9,7,5])
plt.subplot(1,2,2)                       # same index number 
plt.plot(a,b)
```
