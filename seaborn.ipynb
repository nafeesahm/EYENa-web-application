# pairplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue= 'day')
```

```python
sns.pairplot(df, hue= 'sex')
```

# dist plot

```python
sns.distplot(df["total_bill"])               # density at some interval
```

```python
sns.distplot(df['total_bill'], kde= False)                # kde is kernel density estimate
                                                          # what is kernel density estimate :- amount per unit area
```

# works on categorical data

# Box plot

```python
df.head()
```

```python
sns.boxplot(x= 'sex' , y = 'tip' , data=df)            # one numerical data, one categorical data and define data=df
```

# Count plot

```python
sns.countplot('sex' , data =df , hue = 'smoker')                  # hue works for saturate or a kind of filter in data
```

# Bar plot

```python
sns.barplot(x = 'day' , y = 'total_bill' , data = df)
```
